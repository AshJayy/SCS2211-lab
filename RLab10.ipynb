```r
extract_attr <- function(row){
    if(!is.vector(row) && !is.data.frame(row)){
        stop("Input must be a vector or a single-row data frame")
    }
    attributes <- lapply(names(row), function(key) {
        if(key == "Cholesterol"){
            cholesterol_value <- as.numeric(row[[key]]) 
            if(cholesterol_value < 200) {
                cat(key, " : LOW\n")
            }else{
                cat(key, " : HIGH\n")
            }   
        }else{
            cat(key, " : ", row[[key]], "\n")
        }
  })
}
extract_attr(data[1, ])
```
